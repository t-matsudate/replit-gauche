### AtCoder Beginners Selection Practice A

https://atcoder.jp/contests/abs/tasks/practice_1

> 問題文
>
> 高橋君はデータの加工が行いたいです。
> 整数 a, b, c と、文字列 s が与えられます。
> a + b + c の計算結果と、文字列 s を並べて表示しなさい。
>
> 制約
>
> * $1 \le a, b, c \le 1000$
> * $1 \le |s| \le 100$

#### 回答例

```scheme
(let* ((line-1 (read-line (current-input-port)))
       (line-2 (string->split (read-line (current-input-port)) " "))
       (line-3 (read-line (current-input-port)))
       (a (string->number line-1))
       (b (string->number (car line-2)))
       (c (string->number (car (cdr line-2)))))
  (print (+ a b c) " " line-3))
```

#### ポイント

1. コマンドラインからの入力の読み取りは `read-line` で行います。
2. 入力に使う[**ポート**](http://practical-scheme.net/gauche/man/gauche-refj/Ru-Chu-Li-.html#g_t_30dd_30fc_30c8)として `current-input-port` を使います。標準入力は `standard-input-port` ですが、これは `current-input-port` のデフォルト値でもあります。
3. 入力の内、2 行目は空白(`0x20`)区切りで複数存在します。それを b と c にそれぞれ分けるため、 `string->split` で切り離します。ここで、**その返り値はリストである**ことに注意してください。
4. それぞれの入力はまだ文字列であるため、それらを `string->number` で数値に変換します。この時、 2 行目のリストはそれぞれ `car` と `cdr` を使い分けることで個別に処理できます。
5. 最後に、 a, b, c の合計と s を `print` で 1 行に繋げながら出力します。ここで、 `print` は**空白を挟まずに出力する**ことに注意してください。

ちなみに、 Scheme において値と変数を結びつける手段は主に `define`, `let`, `let*`, `letrec`, `letrec*` の 5 つがあります。この内、 `let` と `letrec` は同じスコープ(同じ括弧の範囲内)で定義された変数を他の変数が利用できないため、 `let*` の方を採用しています。
もしくは [内部 `define`](http://practical-scheme.net/gauche/man/gauche-refj/Ding-Yi-.html#g_t_5b9a_7fa9) でもよいです。

### ABC081 A

https://atcoder.jp/contests/abs/tasks/abc081_a

> 問題文
>
> すぬけ君は 1,2,3 の番号がついた 3 つのマスからなるマス目を持っています。各マスには 0 か 1 が書かれており、マス $i$ には $s_i$ が書かれています。
>
> すぬけ君は 1 が書かれたマスにビー玉を置きます。ビー玉が置かれるマスがいくつあるか求めてください。
>
> 制約
>
> * $s_1$, $s_2$, $s_3$ は 1 あるいは 0

#### ポイント

* 条件分岐しながら再帰で文字列(あるいは文字のリスト)を処理していくことになります。
* SICP 1.1.8 までに相当。

再帰は `define` で定義した手続きの中で以下のような書き方をすることで表現できます。

```scheme
(define (f sequence)
  (if (not (null? sequence))
      ; 同じ手続きのどこかで自分自身を呼び直します。
      (f (cdr sequence))))
```

### ABC087 B

https://atcoder.jp/contests/abs/tasks/abc087_b

> 問題文
>
> あなたは、500 円玉を A 枚、100 円玉を B 枚、50 円玉を C 枚持っています。 これらの硬貨の中から何枚かを選び、合計金額をちょうど X 円にする方法は何通りありますか。
>
> 同じ種類の硬貨どうしは区別できません。2 通りの硬貨の選び方は、ある種類の硬貨についてその硬貨を選ぶ枚数が異なるとき区別されます。
>
> 制約
>
> * $0 \le A, B, C \le 50$
> * $A + B + C \ge 1$
> * $50 \le X \le 20000$
> * A, B, C は整数である
> * X は 50 の倍数である

#### ポイント

* 再帰で 3 種類の硬貨それぞれの枚数の範囲内で条件を満たす要素を数え上げます。  
場合によっては末尾再帰が必要になるかもしれません。
* SICP 1.2.1 に相当。

末尾再帰は**手続きの一番深いところ(末尾位置)で自分自身だけを再帰呼び出しをする**書き方です。対して通常の再帰は末尾位置以外での再帰呼び出しや、自分自身以外の手続きを呼び出しながら計算する種類の再帰も含まれています。
